```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://shopee.co.th/')
```

```python
eng_button = driver.find_element_by_xpath('/html/body/div[2]/div[1]/div[1]/div/div[3]/div[2]/button').click()
```

```python
close = driver.execute_script('return document.querySelector("shopee-banner-popup-stateful")\
                                .shadowRoot.querySelector("div.shopee-popup__close-btn")').click()
```

```python
search = driver.find_element_by_xpath('/html/body/div[1]/div/div[2]/div[2]/div/div[1]/div[1]/div/form/input')
```

```python
search.send_keys('Korea dress')
```

```python
search.send_keys(Keys.ENTER)
```

```python
driver.execute_script("document.body.style.zoom='100%'")
```

```python
data = driver.page_source
```

```python
!pip install bs4
```

```python
import bs4
```

```python
soup = bs4.BeautifulSoup(data)
```

```python
product = soup.find_all('div',{'class':'ie3A+n bM+7UW Cve6sh'})
```

```python
all_product = []
for i in product:
    all_product.append(i.text)
all_product
```

```python
price = soup.find_all('div',{'class':'vioxXd rVLWG6'})
```

```python
all_price = []
for x in price:
    all_price.append(x.text)
all_price
```

```python
import pandas as pd
```

```python
Dress_data = pd.DataFrame([all_product,all_price])
```

```python
Dress_data=Dress_data.transpose()
```

```python
Dress_data.columns = (['title','price'])
```
