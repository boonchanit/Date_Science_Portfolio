# Get Data from HTML Website

```python
import requests
import bs4 
import re

def extraction():

    url = 'https://www.leonardoenglish.com/podcasts/how-to-use-podcasts-to-learn-english-like-a-boss'
    
    Data = requests.get(url)
    
    Data.text
    
    soup = bs4.BeautifulSoup(Data.text)
    
    Text = [p.text for p in soup.find_all('p')]
    
    Podcasts = ' '.join(Text)
    
    Podcasts = re.sub('\s',' ',Podcasts)
    return Podcasts
```

```python
podcast = extraction()
podcast
```

## Clean Data

```python
def clean(podcast):
    data = re.sub('[^a-zA-z]',' ',podcast) 
    data = re.sub('\W\s+',' ',data)
    data = re.sub('\s+',' ',data)
    return data
```

```python
data = clean(podcast)
data
```

## Tokenization

```python
import nltk
from nltk.corpus import stopwords
```

```python
def remove_stopword(data):
    
    token = nltk.word_tokenize(data)
    stop_words = set(stopwords.words('english'))
    tokenized = []
    for word in token:
        if word.lower() not in stop_words:
            tokenized.append(word.lower())
    return tokenized
```

```python
no_stopword = remove_stopword(data)
no_stopword
```

```python
nltk.FreqDist(no_stopword)
```
